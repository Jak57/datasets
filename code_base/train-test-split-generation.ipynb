# Loading dataset

```python
import random
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/Jak57/datasets/main/dataverse-2023/previous_trainset_word_ipa_map_37807.csv"
df = pd.read_csv(url)
print("Shape:", df.shape, "\n")
df.head(5)
```

```python
pairs = []
for index, row in df.iterrows():
    word = row['word']
    ipa = row['ipa']
    pairs.append((word, ipa))
print(len(pairs), "pairs loaded..")

random.shuffle(pairs)
train_size = int(len(pairs)*0.95)
train = pairs[:train_size]
test = pairs[train_size:]
print("Length of trainset", len(train), "and length of testset", len(test))
```

# Creating train set

```python
word = []
ipa = []
for t in train:
    word.append(t[0])
    ipa.append(t[1])
print(len(word), len(ipa))
```

```python
print(word[:5])
print(ipa[:5])
```

```python
data = {
    'word': word,
    'ipa': ipa
}

df = pd.DataFrame(data, columns=data.keys())
df.head(5)
```

```python
df.to_csv('prev_dataset_train_35916.csv')
```

# Creating test set

```python
word = []
ipa = []
for t in test:
    word.append(t[0])
    ipa.append(t[1])
print(len(word), len(ipa))
```

```python
print(word[:5])
print(ipa[:5])
```

```python
data = {
    'word': word,
    'ipa': ipa
}

df = pd.DataFrame(data, columns=data.keys())
df.head(5)
```

```python
df.to_csv('prev_dataset_test_1891.csv')
```
